```python
# Import dependencies

import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setup splinter
    
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Scrape Nasa Mars News

nasa_url = "https://redplanetscience.com/"
browser.visit(nasa_url)

# Soupify results
html = browser.html
nasa_soup = bs(html, 'html.parser')

print(nasa_soup.prettify())
```

```python
# Scrape top news article

results = nasa_soup.find('div', class_="list_text")

news_title = results.find('div', class_="content_title").text
news_teaser = results.find('div', class_="article_teaser_body").text

print(news_title)
print(news_teaser)
```

```python
# Scrape JPL Mars Images

jpl_url = 'https://spaceimages-mars.com/'
browser.visit(jpl_url)

# Soupify results
html = browser.html
jpl_soup = bs(html, 'html.parser')

print(jpl_soup.prettify())
```

```python
# Get Mars image link

results = jpl_soup.find('img', class_="headerimage fade-in")["src"]
print(results)
```
